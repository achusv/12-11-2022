```python
df1 = df.iloc[0:7800]
```

```python
df1['Age'] = df1['Age'].str.replace(r'\W',"")
```

```python
df1['Age'] = df1['Age'].str.replace("_","")
df1['Credit_Mix'] = df1['Credit_Mix'].str.replace("_","")
df1['Occupation'] = df1['Occupation'].str.replace("_","")
df1['Amount_invested_monthly'] = df1['Amount_invested_monthly'].str.replace("_","")
df1['Annual_Income'] = df1['Annual_Income'].str.replace("_","")
df1['Num_of_Loan'] = df1['Num_of_Loan'].str.replace("_","")
df1['Num_of_Delayed_Payment'] = df1['Num_of_Delayed_Payment'].str.replace("_","")
df1['Changed_Credit_Limit'] = df1['Changed_Credit_Limit'].str.replace("_","")
df1['Outstanding_Debt'] = df1['Changed_Credit_Limit'].str.replace("_","")
```

```python
df1.head(20)
```

```python
df2 = df1.copy()
df2 = df2.replace(r'^\s*$', float('NaN'),regex = True)
df2.head(10)
```

```python
df2['Age']=df2['Age'].astype(int)
df2['Annual_Income']=df2['Annual_Income'].astype(float)
df2['Num_of_Loan']=df2['Num_of_Loan'].astype(int)
df2['Num_of_Delayed_Payment']=df2['Num_of_Delayed_Payment'].astype(float)
df2['Changed_Credit_Limit']=df2['Changed_Credit_Limit'].astype(float)
df2['Outstanding_Debt'] = df2['Outstanding_Debt'].astype(float)
df2['Amount_invested_monthly']=df2['Amount_invested_monthly'].astype(float)
```

```python
df2.isnull().sum()
```

```python
df2['Name'] = df['Name'].fillna(df['Name'].mode()[0])
df2['Credit_Mix'] = df['Credit_Mix'].fillna(df['Credit_Mix'].mode()[0])
df2['Occupation'] = df['Occupation'].fillna(df['Occupation'].mode()[0])
df2['Monthly_Inhand_Salary'] = df['Monthly_Inhand_Salary'].fillna(df['Monthly_Inhand_Salary'].mode()[0])
df2['Type_of_Loan'] = df['Type_of_Loan'].fillna(df['Type_of_Loan'].mode()[0])
df2['Num_of_Delayed_Payment'] = df['Num_of_Delayed_Payment'].fillna(df['Num_of_Delayed_Payment'].mode()[0])
df2['Num_Credit_Inquiries'] = df['Num_Credit_Inquiries'].fillna(df['Num_Credit_Inquiries'].mode()[0])
df2['Credit_History_Age'] = df['Credit_History_Age'].fillna(df['Credit_History_Age'].mode()[0])
df2['Amount_invested_monthly'] = df['Amount_invested_monthly'].fillna(df['Amount_invested_monthly'].mode()[0])
df2['Credit_Score'] = df['Credit_Score'].fillna(df['Credit_Score'].mode()[0]) 
df2['Monthly_Balance'] = df['Monthly_Balance'].fillna(df['Monthly_Balance'].mode()[0])
df2['Changed_Credit_Limit'] = df['Changed_Credit_Limit'].fillna(df['Changed_Credit_Limit'].mode()[0])
df2['Outstanding_Debt'] = df['Outstanding_Debt'].fillna(df['Outstanding_Debt'].mode()[0])
```
